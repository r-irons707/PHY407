```python
# Pseudocode
# Import  libraries
# Define given timestep and generate time domain
# Define the inital positions for each simulation
# Define the Verlet method in a function where we need
# inital position, and the time vectors to perform the simulation
# Plot the result
# Show x,y plot and the evolution of each component over time
# Show the energy plots for the first simulation, and demonstrate how energy is conserved

# The Verlet function
#   Initialize constants
#   Set up the inital conditions on the position arrays of both particles
#   Compute the first iteration of the velocity components at dt/2
#   Start for loop over len(t)-1
#   Compute the the next position instance from previous mid-point velocity calculation
#   Compute k vector from new position
#   Compute mid-point velocity from new k vector
#   Return the positions vectors


import numpy as np
import matplotlib.pyplot as plt
from MyFunctions import Verlet
from MyFunctions import Lennard_Jones
from matplotlib.ticker import MultipleLocator, AutoMinorLocator

# Define time step 
dt = 0.01
t = np.arange(0,1000*dt,dt)
m = 1

r11 = [4,4]
r21 = [5.2,4]

r12 = [4.5,4]
r22 = [5.2,4]

r13 = [2,3]
r23 = [3.5,4.4]

r1,r2,v1,v2 = Verlet(t,dt,r11,r21)
r3,r4,v3,v4 = Verlet(t,dt,r12,r22)
r5,r6,v5,v6 = Verlet(t,dt,r13,r23)

# plotting mechanics
tdir = 'in'
plt.rcParams['xtick.direction'] = tdir
plt.rcParams['ytick.direction'] = tdir
text = 15
legendfont = 14

# position plot
fig, ((a0,a1,a2),(a3,a4,a5),(a6,a7,a8)) = plt.subplots(figsize=(15,10),ncols=3,nrows=3,\
                                        gridspec_kw={'height_ratios': [3, 1, 1]})
a0.plot(r1[0],r1[1],'.',label="Particle 1",c='k')
a0.plot(r2[0],r2[1],'.',label="Particle 2",c='b',alpha=0.7)
a0.set_title("First Simulation",fontsize=text)
a0.set_xlabel("X-Position",fontsize=text)
a0.set_ylabel("Y-Position",fontsize=text)
a0.legend(loc='best',fontsize=legendfont)
a0.xaxis.set_minor_locator(MultipleLocator(0.1))
a0.yaxis.set_minor_locator(MultipleLocator(0.01))
a0.yaxis.set_ticks_position('both') 
a0.xaxis.set_ticks_position('both')

a1.plot(r3[0],r3[1],'.',label="Particle 1",c='k')
a1.plot(r4[0],r4[1],'.',label="Particle 2",c='b',alpha=0.7)
a1.set_title("Second Simulation",fontsize=text)
a1.set_xlabel("X-Position",fontsize=text)
a1.set_ylabel("Y-Position",fontsize=text)
a1.legend(loc='best',fontsize=legendfont)
a1.xaxis.set_minor_locator(MultipleLocator(1000))
a1.yaxis.set_minor_locator(MultipleLocator(0.01))
a1.yaxis.set_ticks_position('both') 
a1.xaxis.set_ticks_position('both')

a2.plot(r5[0],r5[1],'.',label="Particle 1",c='k')
a2.plot(r6[0],r6[1],'.',label="Particle 2",c='b',alpha=0.7)
a2.set_title("Third Simulation",fontsize=text)
a2.set_xlabel("X-Position",fontsize=text)
a2.set_ylabel("Y-Position",fontsize=text)
a2.legend(loc='best',fontsize=legendfont)
a2.xaxis.set_minor_locator(MultipleLocator(0.1))
a2.yaxis.set_minor_locator(MultipleLocator(0.01))
a2.yaxis.set_ticks_position('both') 
a2.xaxis.set_ticks_position('both')

a3.plot(t,r1[0],c='k')
a3.plot(t,r2[0],c='b',alpha=0.7)
a3.set_ylabel("X-Component",fontsize=text)
a3.set_xlabel("Time [s]",fontsize=text)
a3.xaxis.set_minor_locator(MultipleLocator(1))
a3.yaxis.set_minor_locator(MultipleLocator(0.01))
a3.yaxis.set_ticks_position('both') 
a3.xaxis.set_ticks_position('both')

a4.plot(t,r3[0],c='k')
a4.plot(t,r4[0],c='b',alpha=0.7)
a4.set_ylabel("X-Component",fontsize=text)
a4.set_xlabel("Time [s]",fontsize=text)
a4.xaxis.set_minor_locator(MultipleLocator(1))
a4.yaxis.set_minor_locator(MultipleLocator(5000))
a4.yaxis.set_ticks_position('both') 
a4.xaxis.set_ticks_position('both')

a5.plot(t,r5[0],c='k')
a5.plot(t,r6[0],c='b',alpha=0.7)
a5.set_ylabel("X-Component",fontsize=text)
a5.set_xlabel("Time [s]",fontsize=text)
a5.xaxis.set_minor_locator(MultipleLocator(1))
a5.yaxis.set_minor_locator(MultipleLocator(0.1))
a5.yaxis.set_ticks_position('both') 
a5.xaxis.set_ticks_position('both')

a6.plot(t,r1[1],c='k')
a6.plot(t,r2[1],c='b',alpha=0.7)
a6.set_ylabel("Y-Component",fontsize=text)
a6.set_xlabel("Time [s]",fontsize=text)
a6.xaxis.set_minor_locator(MultipleLocator(1))
a6.yaxis.set_ticks_position('both') 
a6.xaxis.set_ticks_position('both')

a7.plot(t,r3[1],c='k')
a7.plot(t,r4[1],c='b',alpha=0.7)
a7.set_ylabel("Y-Component",fontsize=text)
a7.set_xlabel("Time [s]",fontsize=text)
a7.xaxis.set_minor_locator(MultipleLocator(1))
a7.yaxis.set_ticks_position('both') 
a7.xaxis.set_ticks_position('both')

a8.plot(t,r5[1],c='k')
a8.plot(t,r6[1],c='b',alpha=0.7)
a8.set_ylabel("Y-Component",fontsize=text)
a8.set_xlabel("Time [s]",fontsize=text)
a8.xaxis.set_minor_locator(MultipleLocator(1))
a8.yaxis.set_ticks_position('both') 
a8.xaxis.set_ticks_position('both')

plt.tight_layout()
plt.savefig('Q1a.pdf')
plt.show()

# compute the energies of the two particles for the first simulation
T1 = 0.5*m*v1[0]
T2 = 0.5*m*v2[0]
U1 = Lennard_Jones(r1[0])
U2 = Lennard_Jones(r2[0])
E1 = T1 + U1
E2 = T2 + U2

# energy plots
fig,(ax0,ax1,ax2) = plt.subplots(figsize=(15,10),ncols=1,nrows=3,\
                                        gridspec_kw={'height_ratios': [1, 1, 1]})
ax0.plot(t,T1,c='k',alpha=0.7,label='Particle 1')
ax0.plot(t,T2,c='b',alpha=0.7,label='Particle 2')
ax0.set_title('First simulation kinetic energy',fontsize=text)
ax0.set_xlabel('Time (s)',fontsize=text)
ax0.set_ylabel('Energy',fontsize=text)
ax0.legend(fontsize=text)
ax0.xaxis.set_minor_locator(MultipleLocator(1))
ax0.yaxis.set_minor_locator(MultipleLocator(0.01))
ax0.yaxis.set_ticks_position('both') 
ax0.xaxis.set_ticks_position('both')

ax1.plot(t,U1,c='k',alpha=0.7,label='Particle 1')
ax1.plot(t,U2,c='b',alpha=0.7,label='Particle 2')
ax1.set_title('First simulation potential energy',fontsize=text)
ax1.set_xlabel('Time (s)',fontsize=text)
ax1.set_ylabel('Energy',fontsize=text)
ax1.legend(fontsize=text)
ax1.xaxis.set_minor_locator(MultipleLocator(1))
ax1.yaxis.set_minor_locator(MultipleLocator(0.00001))
ax1.yaxis.set_ticks_position('both') 
ax1.xaxis.set_ticks_position('both')

ax2.plot(t,E1,c='k',alpha=0.7,label='Particle 1')
ax2.plot(t,E2,c='b',alpha=0.7,label='Particle 2')
a2.set_title('First simulation, total energy',fontsize=text)
ax2.set_xlabel('Time (s)',fontsize=text)
ax2.set_ylabel('Energy',fontsize=text)
ax2.legend(fontsize=text)
ax2.xaxis.set_minor_locator(MultipleLocator(1))
ax2.yaxis.set_minor_locator(MultipleLocator(0.01))
ax2.yaxis.set_ticks_position('both') 
ax2.xaxis.set_ticks_position('both')

plt.tight_layout()
plt.savefig('Q1b.pdf')
plt.show()
```
