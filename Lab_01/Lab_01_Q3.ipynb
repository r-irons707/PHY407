```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
from matplotlib.ticker import MultipleLocator, AutoMinorLocator
#text size
text = 15
# we first need an array to act as the size of our matrices that we can loop over
N = 100
N_domain = np.arange(2,N,1)
# create time array to populate
t_mult = []
t_np = []
#start time counter here
for n in range(2,N):
    # configure our two matrices to be operated on
    A = np.ones([n,n],float)*3
    B = np.ones([n,n],float)*3
    
    C_mult = np.zeros([n,n],float)
    C_np = np.zeros([n,n],float)
    # start time counter here
    start_mult = time()
    # start computation as outlined in the text
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i,j] += A[i,k]*B[k,j]
    # end time counter here
    end_mult = time()
    t_mult.append(end_mult - start_mult)
    
    # now timing the np.dot operation
    start_np = time()
    C_np = np.dot(A,B)
    end_np = time()
    t_np.append(end_np - start_np)
    
# plotting the time it takes for matrix multiplication
tdir = 'in'
plt.rcParams['xtick.direction'] = tdir
plt.rcParams['ytick.direction'] = tdir
text = 15

fig, (ax0,ax1) = plt.subplots(ncols=2,nrows=1,figsize=(14,5))
# time plot
ax0.scatter(N_domain,t_mult,color='k',s=8)
ax0.set_title('Time taken for matrix multiplication',fontsize=text)
ax0.set_xlabel('Dimension of matrices (N)',fontsize=text)
ax0.set_ylabel('Time (s)',fontsize=text)
#past this point its fancy plotting mechanics
ax0.xaxis.set_minor_locator(MultipleLocator(2))
ax0.yaxis.set_minor_locator(MultipleLocator(0.05))
ax0.yaxis.set_ticks_position('both') 
ax0.xaxis.set_ticks_position('both')
ax0.grid()
# np.dot plot
ax1.scatter(N_domain,t_np,color='k',s=8)
ax1.set_title('NP.dot time',fontsize=text)
ax1.set_xlabel('Dimension of matrices (N)',fontsize=text)
ax1.set_ylabel('Time (s)',fontsize=text)
#past this point its fancy plotting mechanics
ax1.yaxis.set_ticks_position('both') 
ax1.xaxis.set_ticks_position('both')
ax1.grid()

plt.tight_layout()
plt.savefig('N_time.png')

plt.show()
```
