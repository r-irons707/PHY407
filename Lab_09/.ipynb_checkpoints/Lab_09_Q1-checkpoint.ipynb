```python
import numpy as np
import scipy.optimize as sco
from scipy.constants import hbar
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, AutoMinorLocator
from numpy import copy
from numpy.linalg import inv

# plotting mechanics
tdir = 'in'
plt.rcParams['xtick.direction'] = tdir
plt.rcParams['ytick.direction'] = tdir
text = 18
legendfont = 16
```

```python
def banded(Aa,va,up,down): 
    '''computes a vector x from Ax=v
    Input:
        Aa: contains banded matrix
        va: vector on rhs of equation
        up: how many nonzero elements there are above and below the diagonal, respectively, in each column of the matrix.
    Output:
        v: vector on rhs of equation'''
    A = copy(Aa) 
    v = copy(va) 
    N = len(v) 
    # Gaussian elimination 
    for m in range(N):
        # Normalization factor 
        div = A[up,m]
        # Update the vector first 
        v[m] /= div 
        for k in range(1,down+1): 
            if m+k<N: 
                v[m+k] -= A[up+k,m]*v[m]
        
        # Now normalize and subtract the pivot row 
        for i in range(up): 
            j = m + up - i 
            if j<N: 
                A[i,j] /= div 
                for k in range(1,down+1): 
                    A[i+k,j] -= A[up+k,m]*A[i,j]
    # Backsubstitution 
    for m in range(N-2,-1,-1): 
        for i in range(up): 
            j = m + up - i 
            if j<N: 
                v[m] -= A[i,j]*v[j]
    return v
# Define potential
def V(x):
    return 0
```

```python
# constants
L = 1e-8 # m, length of box
me = 9.109e-31 # kg, mass of electron

# initial conditions
phi0 = 1
x0 = L/5
sigma = L/25
kappa = 500/L

# well conditions
P = 1024
x = np.linspace(-L/2,L/2,P)
arg = -((x - x0)**2/(4*sigma**2) + 1j*kappa*x) # argument for initial condition
phix0t0 = np.exp(arg) # initial condition phi0*

# time span
N = 3000 # time steps
tau = 1e-18 # s, time step size
T = N*tau
time = np.linspace(0,T,N)

# matrices constants for A,B
a = L/N
a1 = 1 + tau*(1j*hbar)/(2*me*a**2)
a2 = -tau*(1j*hbar)/(4*me*a**2)
b1 = 1 - tau*(1j*hbar)/(2*me*a**2)
b2 = tau*(1j*hbar)/(4*me*a**2)

# construct A,B matrices
def trimatrix(a,b,c,d,k1,k2,k3):
    '''defines a tri-diagonal matrix for matrices A,B
    Input:
        a,b,c,d: diagonal values for each entry
        k1,k2,k3: definitions for the diagonal/offdiagonal entries
    Output:
        tri-diagonal matrices A,B'''
    A = np.diag(a,k1) + np.diag(b,k2) + np.diag(b,k3)
    B = np.diag(c,k1) + np.diag(d,k2) + np.diag(d,k3)
    return A,B
n = N-1
a_1 = np.ones(P)*a1; a_2 = np.ones(P-1)*a2; b_1 = np.ones(P)*b1; b_2 = np.ones(P-1)*b2
A,B = trimatrix(a_1,a_2,b_1,b_2,0,-1,1)

# implementation
phi = []
phi.append(phix0t0)
for i in range(len(time)):
    # coding B*psi, first step
    v = np.matmul(B,phi[i])
    # calculate x using banded function
    A_inver = inv(A)
    X = np.matmul(A_inver,v)
    #X = banded(A,v,1,1) # attempted, something aint working here idk why
    phi.append(X)
phi = np.array(phi)
```
