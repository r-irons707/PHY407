```python
import numpy as np
import matplotlib.pyplot as plt

# Import functions from scripts
from dcst_for_q2 import *

def dXXt2(f,xx,yy):
    '''Takes the discrete Fourier transform along x and y (X = C/S)
    input:
        func: 2D array we wish to take the Fourier transform of
        xx,yy: even/odd function direction specifications for the 2D array along the x,y axis directions
                pass in 'even' or 'odd' as strings
    output:
        transf: transformed Fourier of the 2D array'''
    M = f.shape[0] # Number of rows
    N = f.shape[1] # Number of columns
    a = np.zeros((M, N)) # Intermediate array
    b = np.zeros((M, N)) # Final array

    # Take transform along x
    for j in range(N):
        # Check if function is even along this direction
        if xx=="even":
            # Perform DCT
            a[:,j] = dct(f[:,j])
        # Check if function is odd
        elif xx=="odd":
            # Perform DST
            a[:,j] = dst(f[:,j])
    # Take transform along y
    for i in range(M):
        # Check if function is even along this direction
        if yy=="even":
            # Perform DCT
            b[i,:] = dct(a[i,:])
        # Check if function is odd along this direction
        elif yy=='odd':
            # Perform DST
            b[i,:] = dst(a[i,:])
    return b

def idXXt2(b,xx,yy):
    '''Takes the discrete inverse Fourier transform along y and x (X = C/S)
    input:
        transf: 2D array we wish to take the inverse of
        xx,yy: even/odd function direction specifications for the 2D array along the x,y axis directions
                pass in 'even' or 'odd' as strings
    output:
        itransf: inverse transformed Fourier of the 2D array'''
    M = b.shape[0] # Number of rows
    N = b.shape[1] # Number of columns
    a = np.zeros((M, N)) # Intermediate array
    f = np.zeros((M, N)) # Final array
    # inverse transform along yy
    for i in range(M):
        if yy=="even":
            a[i:] = idct(b[i,:])
        elif yy=="odd":
            a[i:] = idst(b[i,:])
    # inverse transform along xx
    for j in range(N):
        if xx=="even":
            f[:,j] = idct(a[:,j])
        elif xx=='odd':
            f[:,j] = idst(a[:,j])
    return f

# define testing domain
x = np.linspace(-5,5,50)
y = np.linspace(-5,5,50)
x, y = np.meshgrid(x, y)

def func_2D(x=0, y=0):
    return x**2 + y**2
function = func_2D(x,y)

f = dXXt2(function,"even","even") # take fourier transform
zprime = idXXt2(f,"even","even") # take inverse fourier transform

fig, (a0,a1,a2) = plt.subplots(figsize=(12,4),ncols=3)
a0.contourf(x,y,function,cmap='inferno')
a0.set_title(r"Original function, $x^2 + y^2$")
a0.set_xlabel("x direction")
a0.set_ylabel("y direction")

a1.imshow(f,origin='lower',cmap='inferno')
a1.set_xlim([0,10])
a1.set_ylim([0,10])
a1.set_title("2D Fourier transform of function")
a1.set_xlabel("Fourier transform in x")
a1.set_ylabel("Fourier transform in y")

a2.contourf(x,y,zprime,cmap='inferno')
a2.set_title(r"Recovered $x^2 + y^2$ function")
a2.set_xlabel("x direction")
a2.set_ylabel("y direction")

plt.tight_layout()
plt.savefig("Q2a_Plot.pdf")
plt.show()

# part b
# constants
Lx, Ly, Jo, m, n, c = 1, 1, 1, 1, 1, 1
P = 32
dx = Lx/P
dy = Ly/P
T = 20
omega = 3.75
dt = 0.01

Ez = np.zeros((P+1,P+1))
Hx = np.zeros((P+1,P+1))
Hy = np.zeros((P+1,P+1))

#discritize t,x,y
t = np.arange(0,T,dt)
x_domain = np.arange(0,Lx+dx,dx)
y_domain = np.arange(0,Ly+dx,dy)
x, y = np.meshgrid(x_domain,y_domain)

def J(x,y,t): # define current density given eq. 7 in lab manual
    '''current density definition with nxnxn dimensions'''
    Jz = []
    for tval in t:
        tmp = Jo*np.sin(m*np.pi*x/Lx)*np.sin(n*np.pi*y/Ly)*np.sin(omega*tval)
        Jz.append(tmp)
    return np.stack(Jz,axis=0)
Jz = J(x,y,t) # define current density

# fourier transform all of Jz
Jzhat_tmp = []
for i in Jz:
    Jzhat_tmp.append(dXXt2(i,xx="odd",yy="odd"))
Jzhat = np.stack(Jzhat_tmp,axis=0)

# fourier transform Hx, Hy and Ez at t=0 (only info we have)
Hxhat = dXXt2(Hx,xx='odd',yy='even')
Hyhat = dXXt2(Hy,xx='even',yy='odd')
Ezhat = dXXt2(Ez,xx='odd',yy='odd')

EzhatS = np.zeros((len(t),P+1,P+1))
HxhatS = np.zeros((len(t),P+1,P+1))
HyhatS = np.zeros((len(t),P+1,P+1))
EzhatS[0] = Ezhat
HxhatS[0] = Hxhat
HyhatS[0] = Hyhat

# implementation
# constants
Dx = np.pi*c*dt/2/Lx
Dy = np.pi*c*dt/2/Ly

Eztmp = []
Hxtmp = []
Hytmp = []
for i in range(len(t)):
    Ezt2 = []
    Hxt2 = []
    Hyt2 = []
    for k in range(len(y_domain)):
        Ezt = []
        Hxt = []
        Hyt = []
        for j in range(len(x_domain)):
            Ei = ((1-x_domain[j]**2*Dx**2 - y_domain[k]**2*Dy**2)*EzhatS[i,k,j] \
                + 2*y_domain[k]*Dy*HxhatS[i,k,j] + 2*x_domain[j]*Dx*HyhatS[i,k,j] + dt*Jzhat[i,k,j])\
                    /(1+x_domain[j]**2*Dx**2 + y_domain[k]*Dy**2)
            Hxi = HxhatS[i,k,j] - y_domain[k]*Dy*(Ei + EzhatS[i,k,j])
            Hyi = HyhatS[i,k,j] - x_domain[j]*Dx*(Ei + EzhatS[i,k,j])
            Ezt.append(Ei)
            Hxt.append(Hxi)
            Hyt.append(Hyi)
        Ezt2.append(Ezt)
        Hxt2.append(Hxt)
        Hyt2.append(Hyt)
    Eztmp.append(Ezt2)
    Hxtmp.append(Hxt2)
    Hytmp.append(Hyt2)
EzHatStack = np.stack(Eztmp,axis=0)
HxHatStack = np.stack(Hxtmp,axis=0)
HyHatStack = np.stack(Hytmp,axis=0)

# inverse fourier to recover values
EzSoltmp = []
HxSoltmp = []
HySoltmp = []
for i in EzHatStack:
    sol = idXXt2(i,xx='odd',yy='odd')
    EzSoltmp.append(sol)
for i in HxHatStack:
    sol = idXXt2(i,xx="odd",yy="even")
    HxSoltmp.append(sol)
for i in HyHatStack:
    sol = idXXt2(i,xx='even',yy='odd')
    HySoltmp.append(sol)
EzSol = np.stack(EzSoltmp,axis=0)
HxSol = np.stack(HxSoltmp,axis=0)
HySol = np.stack(HySoltmp,axis=0)

# sorting values for plotting
x1_keep = int(0.5/dx)
y1_keep = int(0.0/dx)
x2_keep = int(0.0/dx)
y2_keep = int(0.5/dx)
x3_keep = int(0.5/dx)
y3_keep = int(0.5/dx)

# solutions
HxPlotSol = HxSol[:,y2_keep,x2_keep]
HyPlotSol = HySol[:,y1_keep,x1_keep]
EzPlotSol = EzSol[:,y3_keep,x3_keep]

fig,(a0,a1,a2) = plt.subplots(figsize=(12,3),ncols=3)

a0.plot(t,HxPlotSol,c="b",alpha=0.6)
a0.set_xlabel("Time (s)")
a0.set_ylabel(r"H$_x$(x=0.5,y=0)")

a1.plot(t,HyPlotSol,c="k")
a1.set_xlabel("Time (s)")
a1.set_ylabel(r"H$_y$(x=0,y=0.5)")

a2.plot(t,EzPlotSol,c="k",alpha=0.5)
a2.set_xlabel("Time (s)")
a2.set_ylabel(r"E$_z$(x=0.5,y=0.5)")

plt.tight_layout()
plt.savefig("Q2b_Plots.pdf")
plt.show()
```
