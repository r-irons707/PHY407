```python
import numpy as np
from scipy.constants import hbar
from scipy.integrate import simpson
from numpy.linalg import eigvalsh, eigh
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, AutoMinorLocator
```

```python
# stuff for plotting
tdir = 'in'
plt.rcParams['xtick.direction'] = tdir
plt.rcParams['ytick.direction'] = tdir
text = 15
legendfont = 14
```

### Part b

```python
def Hmatrix(n,m):
    # constants
    L = 5e-10 # angstrom to meters
    a = 10*1.60218e-19 # eV to joules
    M = 9.1094e-31 # kg

    test1 = n%2
    test2 = m%2
    if n != m and ((test1 == 0 and test2 == 0) or (test1 != 0 and test2 != 0)):
        # test the m!=n condition and both being either even or odd
        return 0
    
    if n != m and (test1 == 0 and test2 != 0) or (test1 != 0 and test2 == 0):
        # test the m!=n condition and the case where either one is even or odd, but not both
        return -(8*a*m*n)/(np.pi**2*(m**2 - n**2)**2)
    
    if n == m:
        # test the m=n case
        return a/2 + (np.pi**2*hbar**2*m**2)/(2*M*L**2)
    else:
        print('Something is wrong')
        return 0
```

### Part c

```python
mmax = 10
nmax = 10
H = np.zeros((mmax,nmax))
for m in range(1, mmax+1):
    for n in range(1, nmax+1):
        H[m-1, n-1] = Hmatrix(m, n)

evals = eigvalsh(H)/1.60218e-19
print(evals)
```

### Part d

```python
mmaxd = 100 # part d's mmax
nmaxd = 100 # part d's nmax
Hd = np.zeros((mmaxd,nmaxd))

for m in range(1, mmaxd+1):
    for n in range(1, nmaxd+1):
        Hd[m-1, n-1] = Hmatrix(m, n)
evalsd = eigvalsh(Hd)/1.60218e-19
print(evalsd)
```

### Part e

```python
evalsd,evecd = eigh(Hd) # finding eigenvalues and eigenvectors
V0 = evecd.T[0]
V1 = evecd.T[1]
V2 = evecd.T[2]

# constants
L = 5e-10 # angstrom to meters
a = 10*1.60218e-19 # eV to joules
#V = (a*x)/L

psi0 = []
psi1 = []
psi2 = []
domain = np.linspace(0,L,1000)

for x in domain:
    tmp = 0
    for n in range(1,len(V0)-1):
        tmp += V0[n-1]*np.sin((n*np.pi*x)/L)
    psi0.append(tmp)
psi0 = np.array(psi0) # wave function
psi0_prob = np.abs(psi0)**2 # probability density

for x in domain:
    tmp = 0
    for n in range(1,len(V1)-1):
        tmp += V1[n-1]*np.sin((n*np.pi*x)/L)
    psi1.append(tmp)
psi1 = np.array(psi1) # wave function
psi1_prob = np.abs(psi1)**2 # probability density

for x in domain:
    tmp = 0
    for n in range(1,len(V2)-1):
        tmp += V2[n-1]*np.sin((n*np.pi*x)/L)
    psi2.append(tmp)
psi2 = np.array(psi2) # wave function
psi2_prob = np.abs(psi2)**2 # probability density

# normalization for the wave function
A = simpson(psi0_prob)
A = simpson(psi1_prob)
A = simpson(psi2_prob)
psi0_norm = psi0_prob/np.sqrt(A)
psi1_norm = psi1_prob/np.sqrt(A)
psi2_norm = psi2_prob/np.sqrt(A)

fig,ax0 = plt.subplots(figsize=(8,8),ncols=1,nrows=1)
ax0.plot(domain,psi0_norm,color='k',alpha=0.8,label='Normalized ground state')
ax0.plot(domain,psi1_norm,color='r',alpha=0.8,label='Normalized first excited state')
ax0.plot(domain,psi2_norm,color='b',alpha=0.6,label='Normalized second excited state')
ax0.set_title('First three normalized wavefunctions',fontsize=text)
ax0.set_xlabel('Well domain (m)',fontsize=text)
ax0.set_ylabel('Normalized wave function',fontsize=text)
ax0.legend(loc='upper left',fontsize=legendfont)
ax0.set_ylim(-0.001,0.07)
ax0.xaxis.set_minor_locator(MultipleLocator(0.5e-10))
ax0.yaxis.set_minor_locator(MultipleLocator(0.001))
ax0.yaxis.set_ticks_position('both') 
ax0.xaxis.set_ticks_position('both')

plt.savefig('normalized_wavefunctions.pdf')
plt.show()
```
