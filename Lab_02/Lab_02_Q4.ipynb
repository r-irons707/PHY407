#### Part a

```python
tdir = 'in'
plt.rcParams['xtick.direction'] = tdir
plt.rcParams['ytick.direction'] = tdir
text = 15

fig, ax0 = plt.subplots(ncols=1,nrows=1,figsize=(7,5))
# pu plot
ax0.scatter(u,pu,label='p(u) plot',color='k',s=6)
ax0.scatter(u,qu,label='q(u) plot',color='b',s=6)
ax0.set_title(r'$p(u)$ and $q(u)$ plots centered about 1',fontsize=text)
ax0.set_xlabel('Domain about 1',fontsize=text)
ax0.set_ylabel(r'$p(u)$',fontsize=text)
ax0.legend(loc='best')
ax0.grid()

plt.savefig('4a_plot.pdf')
plt.show()
```
