```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from time import time
```

### Part a

```python
# Constants 
M = 100 # Grid squares on a side
xdomain = np.linspace(0,M,101)
ydomain = np.linspace(0,M,101)
v1 = 1.0 # Voltage
v2 = -1.0
target = 1e-6 # Target accuracy
bar1x = np.arange(20,M-19,1)
bar1y = 20
bar2x = np.arange(20,M-19,1)
bar2y = M-19

#Create arrays to hold potential values
phi = np.zeros([M+1,M+1] ,float)
phi[bar1x,bar1y] = v1
phi[bar2x,bar2y] = v2
phiprime = np.empty([M+1,M+1] ,float)

# Main loop 
delta = 1.0

start1 = time()
while delta>target:
    phiprev = phi.copy()
    for i in range(M+1):
        for j in range(M+1):
            if i==0 or i==M or j==0 or j==M:
                phiprime[i,j] = phi[i,j]
            elif (j==bar1y or j==bar2y) and (np.min(bar1x)<=i<=np.max(bar1x)):
                phiprime[i,j] = phi[i,j]
            else:
                phi[i,j] = (phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4
    # Calculate maximum difference from old values
    delta = np.max(np.abs(phi - phiprev)) 

end1 = time()
totaltime1 = end1 - start1
print('Total time for Gauss-Seidel method',totaltime1)

# Make a plot
fig = plt.figure(figsize=(7,7))
plt.imshow(phi)
plt.title('Capacitor model using Gauss-Seidel',fontsize=16)
plt.xlabel('Postion (cm)',fontsize=16)
plt.ylabel('Postion (cm)',fontsize=16)
plt.gray()
plt.savefig('Capacitor model using Gauss-Seidel.pdf')
plt.show()

fig = plt.figure(figsize=(8.5,7))
plt.contourf(xdomain,ydomain,phi)
plt.title('Capacitor model contour plot using Gauss-Seidel',fontsize=16)
plt.xlabel('Postion (cm)',fontsize=16)
plt.ylabel('Postion (cm)',fontsize=16)
plt.colorbar()
plt.savefig('Capacitor model contour plot using Gauss-Seidel.pdf')
plt.show()
```

### Part b

```python
# Constants 
M = 100 # Grid squares on a side
v1 = 1.0 # Voltage
v2 = -1.0
target = 1e-6 # Target accuracy
bar1x = np.arange(20,M-19,1)
bar1y = 20
bar2x = np.arange(20,M-19,1)
bar2y = M-19

# over-relaxation parameters
omega1 = 0.1
omega2 = 0.5

#Create arrays to hold potential values
phi = np.zeros([M+1,M+1] ,float)
phi[bar1x,bar1y] = v1
phi[bar2x,bar2y] = v2
phiprime = np.empty([M+1,M+1] ,float)

# Main loop 
delta = 1.0

start2 = time()
while delta>target:
    phiprev = phi.copy()
    for i in range(M+1):
        for j in range(M+1):
            if i==0 or i==M or j==0 or j==M:
                phiprime[i,j] = phi[i,j]
            elif (j==bar1y or j==bar2y) and (np.min(bar1x)<=i<=np.max(bar1x)):
                phiprime[i,j] = phi[i,j]
            else:
                phi[i,j] = (1+omega1)*(phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4 - phi[i,j]*omega1
                #((1+omega2)*(phi[i+1,j] + phi[i-1,j] + phi[i,j+1] + phi[i,j-1]))/4 - omega2*phi[i,j]
    # Calculate maximum difference from old values
    delta = np.max(np.abs(phi - phiprev)) 

end2 = time()
totaltime2 = end2 - start2
print('Total time for Gauss-Seidel with over-relaxation method',totaltime2)

# Make a plot
fig = plt.figure(figsize=(7,7))
plt.imshow(phi)
plt.title('Capacitor model using Gauss-Seidel with over-relaxation',fontsize=16)
plt.xlabel('Postion (cm)',fontsize=16)
plt.ylabel('Postion (cm)',fontsize=16)
plt.gray()
plt.savefig('Capacitor model using Gauss-Seidel with over-relaxation.pdf')
plt.show()

fig = plt.figure(figsize=(8.5,7))
plt.contourf(xdomain,ydomain,phi)
plt.title('Capacitor model contour plot using Gauss-Seidel with over-relaxation',fontsize=16)
plt.xlabel('Postion (cm)',fontsize=16)
plt.ylabel('Postion (cm)',fontsize=16)
plt.colorbar()
plt.savefig('Capacitor model contour plot using Gauss-Seidel with over-relaxation.pdf')
plt.show()
```
