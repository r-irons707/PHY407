```python
import numpy as np
import matplotlib.pyplot as plt
from time import time


def rhs(r):
    """ The right-hand-side of the equations
    INPUT:
    r = [x, vx, y, vy] are floats (not arrays)
    note: no explicit dependence on time
    OUTPUT:
    1x2 numpy array, rhs[0] is for x, rhs[1] is for vx, etc"""
    M = 10.
    L = 2.

    x = r[0]
    vx = r[1]
    y = r[2]
    vy = r[3]

    r2 = x**2 + y**2
    Fx, Fy = - M * np.array([x, y], float) / (r2 * np.sqrt(r2 + .25*L**2))
    return np.array([vx, Fx, vy, Fy], float)


################################# Q1a ########################################
# %% This next part adapted from Newman's odesim.py --------------------------|
a = 0.0
b = 10.0
N = 1000  # let's leave it at that for now
h = (b-a)/N

tpoints_rk4 = np.arange(a, b, h)
xpoints_rk4 = []
vxpoints_rk4 = []  # the future dx/dt
ypoints_rk4 = []
vypoints_rk4 = []  # the future dy/dt

# below: ordering is x, dx/dt, y, dy/dt
r_rk4 = np.array([1., 0., 0., 1.], float)
# Eq.8.33
for t in tpoints_rk4:
    xpoints_rk4.append(r_rk4[0])
    vxpoints_rk4.append(r_rk4[1])
    ypoints_rk4.append(r_rk4[2])
    vypoints_rk4.append(r_rk4[3])
    k1 = h*rhs(r_rk4)  # all the k's are vectors
    k2 = h*rhs(r_rk4 + 0.5*k1)  # note: no explicit dependence on time of the RHSs
    k3 = h*rhs(r_rk4 + 0.5*k2)
    k4 = h*rhs(r_rk4 + k3)
    r_rk4 += (k1 + 2*k2 + 2*k3 + k4)/6

# adaptive method, redefining variables
delta = 1e-6    # m/s
N = 1000
h = (b-a)/N

# start timing for adaptive method
start = time()

tpoints_original = np.arange(a, b, h)
tpoints_adap = []
x_points_adap = []
vx_points_adap = []
y_points_adap = []
vy_points_adap = []

t = a
r_adap = np.array([1., 0., 0., 1.], float)

while t < b:
    x_points_adap.append(r_adap[0])
    vx_points_adap.append(r_adap[1])
    y_points_adap.append(r_adap[2])
    vy_points_adap.append(r_adap[3])
    tpoints_adap.append(t)
    rho = 0
    
    while rho < 1:
        r_big_step = r_adap.copy()
        r_small_step = r_adap.copy()
        #h_cur = tpoints_adaptive[t] - tpoints_adaptive[t-1]
        
        # h step x(t+h) 2 times (x1)
        for _ in range(2):
            k1 = h * rhs(r_small_step)  # all the k's are vectors
            k2 = h * rhs(r_small_step + 0.5*k1)  # note: no explicit dependence on time of the RHSs
            k3 = h * rhs(r_small_step + 0.5*k2)
            k4 = h * rhs(r_small_step + k3)
            r_small_step += (k1 + 2*k2 + 2*k3 + k4)/6
        
        # 2h step x(t+2h) (x2)
        k1 = 2*h * rhs(r_big_step)
        k2 = 2*h * rhs(r_big_step + 0.5*k1)
        k3 = 2*h * rhs(r_big_step + 0.5*k2)
        k4 = 2*h * rhs(r_big_step + k3)
        r_big_step += (k1 + 2*k2 + 2*k3 + k4)/6
        # estimate error of RK4
        error_x = (r_small_step[0] - r_big_step[0])/30
        error_y = (r_small_step[2] - r_big_step[2])/30
        rho = h*delta / np.sqrt(error_x**2 + error_y**2)
        
        new_h = h * rho**0.25
        
        # prevent new step size goes to infinity
        if new_h > 5*h:
            new_h = 5*h
        h = new_h
    # calculate the next t location
    t = t + 2*h
    
    # Actual RK4 method using updated step (with 2h)
    for _ in range(2):
        k1 = h * rhs(r_adap)  # all the k's are vectors
        k2 = h * rhs(r_adap + 0.5*k1)  # note: no explicit dependence on time of the RHSs
        k3 = h * rhs(r_adap + 0.5*k2)
        k4 = h * rhs(r_adap + k3)
        r_adap += (k1 + 2*k2 + 2*k3 + k4)/6

end = time()
adap_time = end - start
print('The adaptive method took',adap_time,'seconds.')

plt.figure(figsize=(7,7))
plt.plot(xpoints_rk4, ypoints_rk4, 'k:', markersize=1.5, label='Non-adaptive method')
plt.plot(x_points_adap,y_points_adap,'b.',alpha=0.7,markersize=5.5,label='Adaptive method')
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title('Trajectory of a ball bearing around a space rod.', fontsize=16)
plt.axis('equal')
plt.grid()
plt.tight_layout()
plt.legend()
plt.savefig('Q1a.pdf', dpi=150)
```

```python
# part c
# adaptive step
delta = 1e-6    # m/s
N = 1000
h = (b-a)/N

timing_individual = []

tpoints_original = np.arange(a, b, h)
tpoints_adap = []
x_points_adap = []
vx_points_adap = []
y_points_adap = []
vy_points_adap = []

t = a
r_adap = np.array([1., 0., 0., 1.], float)

while t < b:
    x_points_adap.append(r_adap[0])
    vx_points_adap.append(r_adap[1])
    y_points_adap.append(r_adap[2])
    vy_points_adap.append(r_adap[3])
    tpoints_adap.append(t)
    
    rho = 0
    start_each = time()
    while rho < 1:
        r_big = r_adap.copy()
        r_small = r_adap.copy()
        
        # h step x(t+h) 2 times (x1)
        for _ in range(2):
            k1 = h * rhs(r_small)  # all the k's are vectors
            k2 = h * rhs(r_small + 0.5*k1)  # note: no explicit dependence on time of the RHSs
            k3 = h * rhs(r_small + 0.5*k2)
            k4 = h * rhs(r_small + k3)
            r_small += (k1 + 2*k2 + 2*k3 + k4)/6
        
        # 2h step x(t+2h) (x2)
        k1 = 2*h * rhs(r_big)
        k2 = 2*h * rhs(r_big + 0.5*k1)
        k3 = 2*h * rhs(r_big + 0.5*k2)
        k4 = 2*h * rhs(r_big + k3)
        r_big += (k1 + 2*k2 + 2*k3 + k4)/6
        
        error_x = (r_small[0] - r_big[0])/30
        error_y = (r_small[2] - r_big[2])/30
        rho = h*delta / np.sqrt(error_x**2 + error_y**2)
        
        new_h = h * rho**0.25
        
        # Prevent new step size goes to infinity
        if new_h > 5*h:
            new_h = 5*h
        
        h = new_h
    
    # Calculate the next t location
    t = t + 2*h
    
    # Actual RK4 method using updated step (with 2h)
    for _ in range(2):
        k1 = h * rhs(r_adap)  # all the k's are vectors
        k2 = h * rhs(r_adap + 0.5*k1)  # note: no explicit dependence on time of the RHSs
        k3 = h * rhs(r_adap + 0.5*k2)
        k4 = h * rhs(r_adap + k3)
        r_adap += (k1 + 2*k2 + 2*k3 + k4)/6
# calculate total time
    end_each = time()
    timing_individual.append(end_each - start_each)

#%%
plt.figure(figsize=(7,7))
plt.plot(tpoints_adap, timing_individual, '.')
plt.xlabel("$t$ (s)",fontsize=14)
plt.ylabel("Time required to complete adaptive step (s)",fontsize=14)
plt.grid()
plt.xlim([0.5, 10])
plt.ylim([0,0.0001])
plt.title("Size of the time step over time for adaptive method", fontsize=14)

plt.savefig('time_step.pdf')
plt.show()
```
