```python
# libraries
import numpy as np
from matplotlib.pyplot import contourf, xlabel, ylabel, title, colorbar
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, AutoMinorLocator
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# plotting mechanics
tdir = 'in'
plt.rcParams['xtick.direction'] = tdir
plt.rcParams['ytick.direction'] = tdir
text = 15
legendfont = 14
```

```python
SLP = np.loadtxt('SLP.txt')
Longitude = np.loadtxt('lon.txt')
Times = np.loadtxt('times.txt')

fig = plt.figure(figsize=(10,7))

contourf(Longitude, Times, SLP)
xlabel('longitude(degrees)',fontsize=text)
ylabel('days since Jan. 1 2015',fontsize=text)
title('SLP anomaly (hPa)',fontsize=text)
colorbar()

plt.savefig('SLP_contour.pdf')
plt.show()
```

### Part a

```python
# pseudocode for part a
#import and plot the data as a contour map using the intro code
# take the 2D real fourier transform of the SLP data
# separate the m=3 and m=5 case by taking the needed columns, zeroing everything else
# take the 2D real inverse fourier transform of the separated components
# contour plot the results
```

```python
SLP3_fourier = np.fft.rfft2(SLP)
SLP5_fourier = np.fft.rfft2(SLP)
SLP3_fourier[:,:3] = 0
SLP3_fourier[:,4:] = 0
SLP5_fourier[:,:5] = 0
SLP5_fourier[:,6:] = 0

filtered3 = np.fft.irfft2(SLP3_fourier)
filtered5 = np.fft.irfft2(SLP5_fourier)

fig = plt.figure(figsize=(10,7))

contourf(Longitude,Times,filtered3)
xlabel('longitude(degrees)',fontsize=text)
ylabel('days since Jan. 1 2015',fontsize=text)
title('SLP filtered m=3 wave number (hPa)',fontsize=text)
colorbar()
plt.savefig('m=3_separated.pdf')
plt.show()

fig = plt.figure(figsize=(10,7))

contourf(Longitude,Times,filtered5)
xlabel('longitude(degrees)',fontsize=text)
ylabel('days since Jan. 1 2015',fontsize=text)
title('SLP filtered m=5 wave number (hPa)',fontsize=text)
colorbar()
plt.savefig('m=5_separated.pdf')
plt.show()
```
